```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(29)
```

#### First Problem: Show the histogram for a poisson process with 2.5 arrivals per second

### Solution Strategy: 

1) Initialization parameters and creation fo random poisson process with built-in poisson function

2) Plot the distribution

```python
# Initial setup
arrival_r=2.5 # Arrival rate/sec
nNum=100000   # No of random numbers
no_arrivals=np.random.poisson(arrival_r,nNum)  # Calculating the arrivals 
```

```python
plt.hist(no_arrivals,bins=12, density=True)
plt.xlabel('No of Events')
plt.ylabel('Density')
plt.title('Poisson Distribution of arrivals in seconds')
plt.grid()
plt.show()

Expected_val=np.mean(no_arrivals)
print(f" Expected Value: {Expected_val:0.2f}")
variance=np.var(no_arrivals)
print(f" Variance: {variance:0.2f}")
```

The distribution show in the figure has higher distribution around 2.5 which was the arrival rate of our problem. Also, from the calculated expected value, it's evident that the value is also 2.5. Here, the variance and expected value/mean is equal. This also confirms about the property of poisson distribution.



#### Second Problem: Show the histogram for a poisson process with 5 arrivals per hour. Plot it in minutes in the histogram. Also, find the inter-arrival time.

### Solution Strategy: 

1) Initialization parameters and creation fo random poisson process with built-in poisson function 

2) Here, the arrival rate is given in hours. To plot it in minutes, it was divided by 60. 

3) Plot the distribution

4) To find the inter-arrival time,

    a) First, the index of arrivals were selected
    
    b) Then, difference between those indexes were selected
    
    c) Later, a histogram was plotted

```python
arrival_r=5/60 # convert it to minutes 
no_arrivals=np.random.poisson(arrival_r,nNum)
```

```python
plt.hist(no_arrivals,bins=50)
plt.xlabel('No of Events')
plt.ylabel('Counts')
plt.title('Poisson Distribution of arrivals in minutes')
plt.grid()
plt.show()

Expected_val=np.mean(no_arrivals)
print(f" Expected Value: {Expected_val:0.3f}")
variance=np.var(no_arrivals)
print(f" Variance: {variance:0.3f}")

```

The distribution show in the figure has highest distribution between (0 and 0.1) which was near the arrival rate (0.083) of our problem. Also, from the calculated expected value, it's evident that the value is also 0.084. Here, the variance and expected value/mean is also equal. This also confirms about the property of poisson distribution. 

```python
# Inter-arrival time calculation
index=np.where(no_arrivals>0)[0] # Find index where the arrivals were true
difference_time=np.diff(index)   # Find the difference between the arrival indexes
```

```python
plt.hist(difference_time,bins=200, density=True)
plt.xlabel('Time in minutes')
plt.ylabel('Arrival Frequency Normalized')
plt.title('Arrival Time in Minutes')
plt.grid()
plt.show()

Expected_val=np.mean(difference_time)
print(f" Expected Value: {Expected_val:0.3f} Minutes")
variance=np.var(difference_time)
print(f" Variance: {variance:0.3f}")
```

Based on the inter-arrival time plot, it is clear that difference between arrival time would be lower for most of the arrivals. There would be higher interval between arrival times too. But, the frequency of those higher interval would be lower. The inter-arrival times distribution also follows exponential decay curve which is also found in poisson distribution. So, it matches the expected outcome.

Here, the expected value is 12.637 which means that the mean inter-arrival time would be 12.637 minutes. If converted to hours, it would be around 5 arrivals/hour which is close to our given mean arrival rate of 5 in an hour. 

#### Third Problem: In this problem, the arrival rate(lambda) increases linearly from 1 to 10 in 10800 seconds. For each seconds there would be one random number. We need to show the plot with varying lambda. 

### Solution Strategy: 

1) First, vary the arrival rate from 1 to 10 with 10800 steps (1 step= 1 Seconds. So, 10800 seconds would be 10800 steps).

2) Create random numbers using the poisson function for all arrival rates/lambda. 

3) Plot the numbers to see the distribution

```python
arrival_r=np.linspace(1,10,10800) # Lambda from 1 to 10 with 10800 steps
values=np.zeros(10800)            # Create a zero matrix to store all random values
time=10800                        # 3 hours or 10800 seconds mark 
k=1                               # Random number lenght
for i in range(1,time):
    values[i]=np.random.poisson(arrival_r[i],k)   
    
```

```python
plt.hist(values,bins=11,density=True)
plt.xlabel('No of Events')
plt.ylabel('Density')
plt.title('Poisson Distribution of Arrivals with varying Arrival Rate')
plt.grid()
plt.show()

```

##### Explanation of Histogram to Check the Arrival Rate Change Effect: 

First, as the arrival rate linearly varies from 1 to 10. The arrival rate mean would be around 5.5. So, the higher distributions would be centered around 5.5. In the plotted histogram, the higher amount of distribution is found around 5 which is closer to 5.5, it is kind of evident that the arrivals rates are producing a distribution following linear assumption. 

Also, the highest arrival rate here is 10. If the poisson generation is working, then, the variance for arrival rate=10 would also be 10. So, overall distribution would also go towards lowest value around 20 . We are seeing the similar phenomena from the histogram's distribution too. 

These indicates toward the fact that the linearly varying arrival rate is producing expected outcomes. 

Another way of solving could be plotting the values with respect to time. 

 

```python
times=np.linspace(0,3,10800) # Time in seconds (0 to 3 hours)
plt.plot(times,values)       # Plot to show events wrt to time
plt.ylabel('Events')
plt.xlabel('Time in Hours')
plt.title('Poisson Distribution of Arrivals with varying Arrival Rate')
plt.grid()
plt.show()
```

The plot shows that with increased arrival rate the mean/center of distribution is increasing too. At 3 hour point, the distribution is centered to 10. As the distribution center is shifting and the wdith of distribution is increasing too, we can say that the change in arrival rate is working.
